```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(47)
x = np.random.normal(size = 500)
error = np.random.normal(scale = 0.6, size = 500)
y = -x**3 + error
plt.plot(x, y, '.')
plt.plot(np.sort(x), -np.sort(x)**3, color = "red")
```

```python
df = pd.DataFrame({"x": x, "y":y})
print("================== Pearson ===============")
print(df.corr())
print("================== Spearman ===============")
print(df.corr(method = "spearman"))
df.corr()
print("================== Kendall ===============")
print(df.corr(method = "kendall"))
```

```python
rho = 0.808824

matrix = [[ 1, rho],
          [rho, 1]]

np.random.seed(47)
X = np.random.multivariate_normal(mean = [0, 0], cov = matrix, size = 500)

plt.plot(X[:,0], X[:,1], '.')

print("================== Pearson ===============")
print(pd.DataFrame(X).corr())
print("================== Spearman ===============")
print(pd.DataFrame(X).corr(method = "spearman"))
```

```python
import numpy as np
np.arange(-1, 1.05, 0.05)
```

```python
np.random.seed(47)
corr_pearson = []
corr_spearman = []
corr_kendall = []
for rho in np.arange(-1, 1.05, 0.05):
  matrix = [[  1, rho],
          [rho,   1]]

  X = np.random.multivariate_normal(mean = [0, 0], cov = matrix, size = 1000)
  df = pd.DataFrame(X)
  corr_pearson.append(df.corr().values[0, 1])
  corr_spearman.append(df.corr(method = "spearman").values[0, 1])
  corr_kendall.append(df.corr(method = "kendall").values[0, 1])
```

```python
plt.plot(np.arange(-1, 1.05, 0.05), corr_pearson, '.', label = "Pearson")
plt.plot(np.arange(-1, 1.05, 0.05), corr_spearman, label = "Spearman")
plt.plot(np.arange(-1, 1.05, 0.05), corr_kendall, label = "Kendall")
plt.xlabel("Rho original")
plt.ylabel("Rho estimado")
plt.title("Correlación")
plt.legend()
```

```python
plt.plot(np.arange(-1, 1.05, 0.05), corr_pearson, '.', label = "Pearson")
plt.xlabel("Rho original")
plt.ylabel("Correlación estimado (Pearson)")
plt.title("Pearson")
plt.figure()
plt.plot(np.arange(-1, 1.05, 0.05), corr_pearson, '.', label = "Spearman")
plt.xlabel("Rho original")
plt.ylabel("Correlación estimado (Spearman)")
plt.title("Spearman")
plt.figure()
plt.plot(np.arange(-1, 1.05, 0.05), corr_pearson, '.', label = "Kendall")
plt.xlabel("Rho original")
plt.ylabel("Correlación estimado (Kendall)")
plt.title("Kendall")
plt.legend()
```
